```python
def eliminar_palabras_repetidas(texto):
    # Dividir el texto en una lista de palabras
    lista_palabras = texto.split()
    # Crear un diccionario para realizar un seguimiento de las palabras únicas
    diccionario_palabras = {}
    for palabra in lista_palabras:
        # Si la palabra ya está en el diccionario, ignorarla
        if palabra in diccionario_palabras:
            continue
        # Si la palabra no está en el diccionario, agregarla como clave con un valor de 1
        diccionario_palabras[palabra] = 1
    # Crear una nueva lista a partir de las claves del diccionario
    lista_sin_repetidos = list(diccionario_palabras.keys())
    # Crear un nuevo texto a partir de la lista sin palabras repetidas
    nuevo_texto = ' '.join(lista_sin_repetidos)
    return nuevo_texto
```

```python
def datosPlazas(nombre):

        #print( 'Plaza: ' + nombre)
        busqueda = nombre.replace(" ", "-").lower()
        a,b = 'áéíóúü','aeiouu'
        trans = str.maketrans(a,b)
        lugar = "quintana-roo"
        
        url = f"https://www.mallsmexico.com/plazas/{lugar}/{busqueda.translate(trans)}"
        #print(url)
        browser = uc.Chrome() 
        browser.get(url)
        html = browser.page_source
        soup = bs(html, 'lxml')
        
        plaza_nombre = soup.find("li", {"class": "active"}).text
        print(plaza_nombre)
        #horario = soup.find("div", {"class": "ul-horario-1"}).find("p").text
        
        espici = soup.find("div", {"class": "ul-beneficios"}).text.replace("\n", "").replace("  ", ",")[1:]
        
        niveles = soup.find("div", {"class": "col-sm-8 pleca-gris-interna"}).find("p").text
        
        #soup.find("div", {"id": "div-accordion-tiendas"}).find_all("img", alt = True)
        # aqui denro estan las categorias y tiendas

        # Aqui estan las categorias de la plaza
        xd = []
        for foo in soup.find("div", {"id": "div-accordion-tiendas"}).find_all("img", alt = True):
            #print(foo['alt'])
            #print(foo)
            xd.append(foo['alt'])
        #print(xd)

        
        tiendas = soup.find("div", {"id": "div-accordion-tiendas"}).find_all("p", {"class":"plugar"})
        print(tiendas)
        todasTiendas = []
        for tienda in tiendas:
            #print(tienda.text)
            todasTiendas.append(tienda.text)
            #print(todasTiendas)
        
        #datos["horario"] = horario
        plazas["nombre"] = nombre
        plazas["espicificaiones"] = espici
        plazas["niveles"] = niveles
        plazas["categorias"] = xd
        plazas["Tiendas"] = todasTiendas
        
        #df = df.append({'first_name': 'Josy', 'last_name':'Clarae', 'gender':'Female'}, ignore_index=True)
        #df = df.append({'Nombre':nombre, 'Niveles':espici, 'Especificaciones':niveles, 'Categorias':xd, 'Tiendas': todasTiendas}, ignore_index=True)
        
        df_plazas = pd.DataFrame(plazas)
        
        return(df_plazas.T)
        
             
```
